# Training Models

## Linear Regression

```python
import numpy as np
```

```python
X = 2* np.random.rand(100,1)
y = 4*3 + np.random.randn(100,1)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

```python
LR.fit(X,y)
X_new = np.array([[0], [2]])
```

```python
print(LR.intercept_)
print(LR.coef_)
pred = LR.predict(X_new)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y,color = "red")
plt.plot(pred)
```

## Gradient Decent

```python
from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor(max_iter= 1000)
sgd.fit(X,y.ravel())
```
